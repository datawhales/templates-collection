```python
def one_hot_encoding(df: pd.DataFrame, column_name: str) -> pd.DataFrame:
    from sklearn.preprocessing import OneHotEncoder

    ohe = OneHotEncoder(sparse=False)

    # One hot encoding to "Grade" column
    one_hot_encoded = ohe.fit_transform(df[column_name].values.reshape(-1, 1))

    # One hot encoded dataframe
    new_df = pd.DataFrame(one_hot_encoded, columns=[f"{column_name}_{x}" for x in ohe.categories_[0]])

    return pd.concat([df.drop(columns=column_name), new_df], axis=1)
```

```python
one_hot_encoding(train_df, "Sex")
```
